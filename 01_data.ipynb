```python
import pandas as pd
```

```python
df_verano = pd.read_csv('data/planilla-verano-2-2023.csv')
df_verano
```

```python
for i in range(9, 19):
  df_verano = df_verano.drop(f'Unnamed: {i}', axis=1)
df_verano = df_verano.rename(columns={'CodSis':'codsis', 'Nombre':'nombre'})
df_verano = df_verano.drop('Num.', axis=1)
df_verano['curso'] = ['verano' for i in range(1, 184)]
df_verano
```

```python
df_invierno = pd.read_csv('data/planilla-invierno-3-2023.csv')
df_invierno
```

```python
for i in range(9, 21):
  df_invierno = df_invierno.drop(f'Unnamed: {i}', axis=1)
df_invierno = df_invierno.rename(columns={'Código':'codsis', 'Alumno':'nombre', 'Carrera': 'carrera'})
df_invierno = df_invierno.drop('Nro', axis=1)
df_invierno['curso'] = ['invierno' for i in range(1, 234)]
df_invierno
```
